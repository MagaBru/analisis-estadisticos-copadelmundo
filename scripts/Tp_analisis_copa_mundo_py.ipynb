```python
import pandas as pd
import os
os.makedirs('resultados', exist_ok= True)
ruta= 'datos/copas_del_mundo.csv'
datos= pd.read_csv(ruta, sep= '\t')

#---- 1) Promedio de goles por partido -----
#Sumamos los goles de para obtener total y luego se divide por la
#cantidad de partidos jugados
total_goles= datos['home_score'].sum() + datos['away_score'].sum()
promedio_goles= total_goles / len(datos)
print (f"Promedio de goles por partido: {promedio_goles:.2f}")

#---- 2) Partidos ganados por equipo ----
datos['ganador']= 'Empate'
datos.loc[datos['home_score']> datos['away_score'], 'ganador']= datos['home']
datos.loc[datos['away_score']> datos['home_score'], 'ganador'] = datos ['away']

partidos_ganados= datos[datos['ganador'] != 'Empate']['ganador'].value_counts()
print ("Partidos ganadospor equipo - Top 5: ")
print (partidos_ganados.head())

#---- 3) Tabla de posiciones ----
# Una victoria suma 3 puntos, un empate 1 y una derrota 0
puntos_home= datos[['home', 'home_score', 'away_score']].copy()
puntos_home.columns= ['equipo', 'goles_favor', 'goles_contra']
puntos_home ['puntos']= puntos_home.apply(lambda x: 3 if x['goles_favor']>x['goles_contra'] else (1 if x['goles_favor']== x['goles_contra']else 0), axis=1)

puntos_away= datos[['away', 'away_score', 'home_score']].copy()
puntos_away.columns= ['equipo', 'goles_favor', 'goles_contra']
puntos_away ['puntos']= puntos_away.apply(lambda x: 3 if x['goles_favor']>x['goles_contra'] else (1 if x['goles_favor']== x['goles_contra']else 0), axis=1)

tabla_posiciones= pd.concat([puntos_home, puntos_away]).groupby('equipo')['puntos'].sum().sort_values(ascending= False)
# Se guardan los resultados
partidos_ganados.to_csv('resultados/partidos_ganados.csv')
tabla_posiciones.to_csv('resultados/tabla_posiciones.csv')

#---- IMPRESIÓN DE RESULTADOS ----
print ("Partidos ganados por equipo: Top 5: ")
print (partidos_ganados.head().to_string())

# Tabla de posiciones
print("Tabla de posiciones: ")
print (tabla_posiciones.to_string())

# Resumen estadístico
print ("Resumen estadístico: ")
with open ('resultados/resumen_estadistico.txt', 'w') as f:
   f.write(f"Promedio de goles por partido: {promedio_goles:.2f}")
   f.write("\n")
   f.write("Partidos ganados por equipo: Top 5: ")
   f.write(partidos_ganados.head().to_string())
   f.write("\n")
print ("Análisis completado y guardado en la carpeta resultados")
```
